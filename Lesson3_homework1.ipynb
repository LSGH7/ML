**В результате группировки видим, что почти все значения среднего, минимума и максимума параметров у злокачественных новообразований значительно (в 1,5 - 2 раза) больше, чем у доброкачественных. Не соответствуют этой тенденции наибольшие значения гладкости и худшей вогнутости (concavity_worst), которые у злокачественных даже чуть меньше, чем у доброкачественных.**

## 3. Гистограммы и корреляции.

```python
def histogram(data1, data2, title="", data1_name="", data2_name="", data1_color="#539caf", data2_color="#495d97", x_label="", y_label="", n_bins = 25):
    _, ax = plt.subplots()
    ax.hist(data1, bins = n_bins, color = data1_color, alpha = 0.75, label = data1_name)
    ax.hist(data2, bins = n_bins, color = data2_color, alpha = 0.75, label = data2_name)
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title)
    
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, labels)
    ax.legend(loc = 'best')
```

```python
# пары признаков с корреляцией больше 0.9
CorField = []
for i in Corr:
    for j in Corr.index[Corr[i] > 0.9]:
        if i != j and j not in CorField and i not in CorField:
            CorField.append(j)
            print (i, j, Corr[i][Corr.index == j].values[0])
```
